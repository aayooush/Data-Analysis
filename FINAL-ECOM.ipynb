```python
import numpy as np
import csv
import matplotlib.pyplot as plt
from random import *
```

```python
#opening all csv files
with open("January.csv", 'r') as j:
    l1 = list(csv.reader(j, delimiter=","))
    
with open("February.csv", 'r') as f:
    l2 = list(csv.reader(f, delimiter=","))

with open("March.csv", 'r') as m:
    l3 = list(csv.reader(m, delimiter=","))

with open("April.csv", 'r') as a:
    l4 = list(csv.reader(a, delimiter=","))
    
with open("May.csv", 'r') as ma:
    l5 = list(csv.reader(ma, delimiter=","))

with open("June.csv", 'r') as ju:
    l6 = list(csv.reader(ju, delimiter=","))
```

```python
#-----JANUARY------#

Jan = np.array(l1)#Array convert

JanCountry = list(Jan[1:,2])#repeated countries
JanCont = list(set(JanCountry))#convert to list wiith no repetition
JanCont.sort()#sort list 1 baar entry

x1 = JanCont
y1 = [JanCountry.count(i) for i in JanCont]

JanCity = list(Jan[1:,1]) #fetch cities list
JanCit = list(set(JanCity)) #convert to list with no repetition
JanCit.sort() #sort list

n1 = JanCit
m1 = [JanCity.count(i) for i in JanCit]

#-----FEBRUARY------#

Feb = np.array(l2)#Array convert

FebCountry = list(Feb[1:,2]) #fetch country list
FebCont = list(set(FebCountry)) #convert to list with no repetition
FebCont.sort() #sort list

x2 = FebCont
y2 = [FebCountry.count(i) for i in FebCont]

FebCity = list(Feb[1:,1])
FebCit = list(set(FebCity)) #convert to list
FebCit.sort() #sort list

n2 = FebCit
m2 = [FebCity.count(i) for i in FebCit]

#-----MARCH------#

Mar = np.array(l3)#Array convert

MarCountry = list(Mar[1:,2])
MarCont = list(set(MarCountry))#convert to list
MarCont.sort()#sort list

x3 = MarCont
y3 = [MarCountry.count(i) for i in MarCont]

MarCity = list(Mar[1:,1])
MarCit = list(set(MarCity))#convert to list
MarCit.sort()#sort list

n3 = MarCit
m3 = [MarCity.count(i) for i in MarCit]

#-----APRIL------#

Apr = np.array(l4)#Array convert

AprCountry = list(Apr[1:,2])
AprCont = list(set(AprCountry))#convert to list
AprCont.sort()#sort list

x4 = AprCont
y4 = [AprCountry.count(i) for i in AprCont]

AprCity = list(Apr[1:,1])
AprCit = list(set(AprCity))#convert to list
AprCit.sort()#sort list

n4 = AprCit
m4 = [AprCity.count(i) for i in AprCit]

#-----MAY------#

May = np.array(l5)#Array convert

MayCountry = list(May[1:,2])
MayCont = list(set(MayCountry))#convert to list
MayCont.sort()#sort list

x5 = MayCont
y5 = [MayCountry.count(i) for i in MayCont]
    
MayCity = list(May[1:,1])
MayCit = list(set(MayCity))#convert to list
MayCit.sort()#sort list

n5 = MayCit
m5 = [MayCity.count(i) for i in MayCit]

#-----JUNE------#

Jun = np.array(l6)#Array convert

JunCountry = list(Jun[1:,2])
JunCont = list(set(JunCountry))#convert to list
JunCont.sort()#sort list

x6 = JunCont
y6 = [JunCountry.count(i) for i in JunCont]

JunCity = list(Jun[1:,1])
JunCit = list(set(JunCity)) #convert to list
JunCit.sort() #sort list

n6 = JunCit
m6 = [JunCity.count(i) for i in JunCit]
```

```python
Cities = JanCity + FebCity + MarCity + AprCity + MayCity + JunCity
Countries = JanCountry + FebCountry + MarCountry + AprCountry + MayCountry + JunCountry
Country = list(set(Countries))
Country.sort()
City = list(set(Cities))

Data = { i : Cities.count(i) for i in City}
Data2 = { Cities[i] : Countries [i] for i in range(len(Cities))}

DataSet = {i : {j : Data[j] for j in City if Data2[j] == i} for i in Countries}
```

```python
Colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

for i in Country:
    plt.figure(figsize=(2,2))
    x = list(DataSet[i].keys())
    y = list(DataSet[i].values())
    shuffle(Colors)
    plt.pie(y,labels=x,colors=Colors)
    plt.title(i+"     TOTAL SALES="+str(sum(y)))
    plt.show() 
```

```python
Month = ["1 Jan" , "2 Feb" , "3 Mar" , "4 Apr" , "5 May" , "6 Jun"]

DataAll = []
for i in range(len(Country)):
    DataAll.append({Month[0] : y1[i], Month[1] : y2[i] ,Month[2]  : y3[i] ,Month[3] : y4[i] ,Month[4] : y5[i] ,Month[5] : y6[i]})
```

```python
colour = ['red','green','blue','gray','gold']

for i in range(len(DataAll)):
    x = list(DataAll[i].keys())
    y = list(DataAll[i].values())
    plt.figure(figsize=(3.5,2))
    plt.plot(x,y,color=colour[i%5])
    plt.title(Country[i])
    plt.show()
    
```
